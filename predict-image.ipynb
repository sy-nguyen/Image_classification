```python
import numpy as np
import matplotlib.pyplot as plt

from urllib.request import urlopen
from PIL import Image
from keras.models import load_model
from tensorflow.keras.applications.resnet50 import preprocess_input
from tensorflow.keras.models import Model
from tensorflow.keras.preprocessing import image
```

```python
model_path = 'model/model_resnet50.h5'

try:
    model = load_model(model_path)
    print("Model loaded successfully.")
except Exception as e:
    print(f"Error loading model: {e}")

labels = ['Cell Phones and Accessories', 'Clothing Men', 'Electronics', 'Shoes', 'Watches']
```

```python
def predict_label_from_url(url):
    try:
        response = urlopen(url)
        img = Image.open(response)
        plt.imshow(img)
        plt.axis('off')
        plt.show()

        img = img.resize((224, 224))
        img = img.convert('RGB')
        img_arr = np.array(img).astype(np.float32)
        img_arr = preprocess_input(img_arr)
        img_arr = np.expand_dims(img_arr, axis=0)
        
        predictions = model.predict(img_arr, verbose=0)
        predicted_class = np.argmax(predictions, axis=1)[0]
        predicted_label = labels[predicted_class]

        print(f"Predicted label: {predicted_label}")

    except Exception as e:
        print(f"Error loading image from URL: {e}")
```

```python
url = input("Import URL image: ")
predict_label_from_url(url)
```
